```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
time = [0, 3, 6, 9, 12, 15, 18, 21, 24]
temperature = [20, 19, 18, 23, 29, 31, 28, 23, 20]
plt.figure(figsize=(10, 2))
plt.xlim(-1,25)
plt.ylim(15,35)
plt.xticks(time)
plt.plot(time,temperature,color = "orange",linestyle = "solid",linewidth = 3.0)
plt.title("5/31 Tokyo Temperature")
plt.xlabel("Time")
plt.ylabel("℃")
plt.fill_between(time,temperature,17,color="lightyellow")

for x,y in zip(time,temperature):
    plt.text(x-0.3,y+1,y,color="gray")

plt.show()
```

```python
plt.bar(time,temperature,color = "lightyellow",edgecolor = "gold",width = 2.0,linestyle = "-",linewidth = 2.0)
plt.show()
plt.barh(time,temperature,color = 'lightyellow',edgecolor = 'gold',height = 2.0,linestyle = '-',linewidth=2.0)
plt.show()
plt.scatter(time, temperature, s=50, color='lightyellow', edgecolor='gold', linewidth=2.0)
plt.show()

data = [40, 25, 15, 20]
lang = ["Python", "Javascript", "C++", "Others"]
plt.pie(data, labels=lang, counterclock = False, startangle=90)
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.interpolate import interp1d

time = [0, 3, 6, 9, 12, 15, 18, 21, 24]
temperature = [20, 19, 18, 23, 29, 31, 28, 23, 20]

f = interp1d(time,temperature,kind="cubic")
x_new = np.linspace(0,24,num=200)

plt.figure(figsize=(10, 2))
plt.xlim(-1,25)
plt.ylim(15,35)
plt.xticks(time)
plt.plot(x_new,f(x_new),color = "gold",linestyle = "solid",linewidth = 2.0)
plt.title("5/31 Tokyo Temperature")
plt.xlabel("Time")
plt.ylabel("℃")
plt.fill_between(x_new,f(x_new),17,color="lightyellow")

for x,y in zip(time,temperature):
    plt.text(x-0.3,y+1,y,color="gray")

plt.tick_params(labelbottom = True,labelleft = False)
plt.tick_params(bottom = False,left = False)
plt.box(False)
    
plt.show()
```

```python
x = np.array([0, 1, 2, 3])
y = np.array([-1, 1, -1, 1])
f = interp1d(x, y, kind="cubic")
x_new = np.linspace(0, 3, num=100)
plt.figure(figsize=(10, 2))
plt.plot(x_new, f(x_new), color='gold',  linestyle='solid', linewidth = 2.0, label='interpoluted')
plt.plot(x, y, color='skyblue', linestyle='solid', linewidth=2.0, label='raw data')
plt.legend()
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax2 = ax1.twinx()
p1 = ax1.plot([0,1,2],[0,1,2],label = "Density",color = "green")
p2 = ax2.plot([0,1,2],[0,3,2],label = "Temperature",color = "deeppink")
leg = fig.legend()
ax1.set_xlabel("Distance")
ax1.set_ylabel("Density")
ax2.set_ylabel("Temperature")
ax1.yaxis.get_label().set_color(p1.get_color())
ax2.yaxis.get_label().set_color(p2.get_color())
leg.texts[0].set_color(p1.get_color())
leg.texts[1].set_color(p2.get_color())

plt.show()
```

```python
import matplotlib.ticker as ticker

fig = plt.figure(figsize = (10,10))
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)

ax1.xaxis.set_major_locator(ticker.MultipleLocator(0.5*np.pi))
ax1.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: f'{np.rad2deg(x)}'))
ax1.yaxis.set_major_locator(ticker.MultipleLocator(0.5))
ax1.yaxis.set_minor_locator(ticker.MultipleLocator(0.25))

ax2.xaxis.set_major_locator(ticker.MultipleLocator(0.5*np.pi))
ax2.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: f'{x/np.pi}$\pi$'))
ax2.yaxis.set_major_locator(ticker.MultipleLocator(0.5))
ax2.yaxis.set_minor_locator(ticker.MultipleLocator(0.25))

ax1.grid(which='major',color='black',linestyle='-')
ax1.grid(which='minor',color='gray',linestyle=':')

ax2.grid(which='major',color='black',linestyle='-')
ax2.grid(which='minor',color='gray',linestyle=':')

x = np.linspace(-2 * np.pi,2 * np.pi,200)
ax1.plot(x,np.sin(x),color = "lightblue",linewidth = "3.0",label = "sin(x)")
ax1.plot(x,np.cos(x),color = "pink",linewidth = "3.0",label = "cos(x)")

ax2.plot(x,np.sin(x),color = "lightblue",linewidth = "3.0",label = "sin(x)")
ax2.plot(x,np.cos(x),color = "pink",linewidth = "3.0",label = "cos(x)")

ax1.legend(loc = "lower left",title = "legend",labelcolor = "linecolor")
ax2.legend(loc = "lower left",title = "legend",labelcolor = "linecolor")
# sin: color='lightblue', linewidth="3.0"
# cos: color="pink", linewidth="3.0"
```
